
# Statistical Business Analysis

This notebook performs:
- Descriptive Statistics
- Distribution Analysis
- Correlation Analysis
- Hypothesis Testing
- Confidence Intervals
- Regression Analysis


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv('business_data.csv')
df.head()
```

```

# Descriptive Statistics
desc_stats = df[['Quantity','Price','Total_Sales','Marketing_Spend']].describe()
desc_stats
```

```

# Histogram
df['Total_Sales'].hist(bins=15)
plt.title('Sales Distribution')
plt.show()
```

```

# Correlation Matrix
corr = df[['Quantity','Price','Total_Sales','Marketing_Spend']].corr()
corr
```

```

# Hypothesis Tests
t1 = stats.ttest_1samp(df['Total_Sales'], popmean=100000)
t2 = stats.ttest_ind(
    df[df['Region']=='East']['Total_Sales'],
    df[df['Region']=='West']['Total_Sales'],
    equal_var=False
)

groups = [grp['Total_Sales'].values for _, grp in df.groupby('Product')]
anova = stats.f_oneway(*groups)

print(t1)
print(t2)
print(anova)
```

```

# Confidence Interval
mean_sales = df['Total_Sales'].mean()
std_sales = df['Total_Sales'].std()
n = len(df)

margin_error = 1.96 * (std_sales / np.sqrt(n))

(mean_sales - margin_error, mean_sales + margin_error)
```

```

# Regression Analysis
X = df[['Marketing_Spend']]
y = df['Total_Sales']

model = LinearRegression()
model.fit(X, y)

pred = model.predict(X)

r2 = r2_score(y, pred)

print('Coefficient:', model.coef_[0])
print('Intercept:', model.intercept_)
print('R-squared:', r2)
```
